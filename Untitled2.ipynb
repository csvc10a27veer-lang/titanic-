```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Visualization style
sns.set(style="whitegrid")
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
netflix = pd.read_csv("netflix_titles.csv")
netflix.head()
```

```python
iris.info()
iris.describe()
iris.isnull().sum()
```

```python
netflix.info()
netflix.describe(include='all')
netflix.isnull().sum()
```

```python
plt.figure(figsize=(12,8))

iris[['sepal_length','sepal_width','petal_length','petal_width']].hist(
    bins=20, figsize=(12,8)
)

plt.suptitle("Iris Feature Distributions")
plt.show()
```

```python
sns.countplot(x='species', data=iris)
plt.title("Count of Iris Species")
plt.show()
```

```python
sns.countplot(x='type', data=netflix)
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='species', y='petal_length', data=iris)
plt.title("Petal Length by Species")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(iris.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap - Iris Dataset")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(y='rating', data=netflix, order=netflix['rating'].value_counts().index)
plt.title("Netflix Content by Rating")
plt.show()
```

```python
netflix['release_year'].value_counts().sort_index().plot(figsize=(12,5))
plt.title("Netflix Content Growth Over Years")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```
